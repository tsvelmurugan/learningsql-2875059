```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.master("local[*]").appName("Colab PySpark").getOrCreate()
```

```python
print(spark.version)
```

```python
data=[("Ram",40,100000),("Dasvid",41,120000),("Ravi",25,250000),("Bala",35,180000)]
colmn=["Name","Age","Salary"]
```

```python
df=spark.createDataFrame(data,colmn)
```

```python
print(str(df.schema))
```
